```python
import glob
import pandas as pd
import numpy as np
import pandas as pd
import scipy as scp
import sklearn
import itertools
from scipy.optimize import fsolve
from upsetplot import generate_data, plot, from_memberships
from collections import Counter
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import StrMethodFormatter
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.patches as mpatches
import matplotlib.ticker as ticker

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
%config InlineBackend.figure_format = 'retina'
```

```python
fig, ax1 = plt.subplots(figsize=(10, 8))

L = np.linspace(4, 14, 200)

dg = 100
n=4**L

ax1.plot(L,np.log(1-dg/n)/np.log(1-1/n) - dg, color="black", linewidth=3)
ax1.set_xlabel('Length of UMI')
ax1.set_ylabel("Counts lost in one gene (out of 100)")
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.savefig("theoretical_counts_lost.pdf")
plt.show()
```
